```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dataset_telemetry.csv")
```

```python
df.head()
```

```python
plt.hist(df["gender"])
```

```python
df_city = df.groupby('city').count().reset_index()
df_city
```

```python
plt.figure(figsize=(15, 5)) 
plt.bar(df_city['city'], df_city['value'], width=0.5)
plt.xticks(df_city['city'],df_city['city'], rotation='vertical')
```

```python
df_category = df.groupby('category').agg({"value":"sum","userid":"count"}).reset_index().sort_values(by="value")
df_category
```

```python
plt.figure(figsize=(15, 5)) 
plt.bar(df_category['category'], df_category['value'], width=0.5)
# plt.xticks(df_category['city'],df_category['city'], rotation='vertical')
```

```python
plt.figure(figsize=(15, 5)) 
plt.bar(df_category['category'], df_category['userid'], width=0.5)
```
