```python
import pandas as pd
data = pd.read_csv("dataset_telemetry.csv")
```

```python
data
```

```python
data = data.drop_duplicates(subset=["timestamp", "userid"]) # Удалим повторяющиеся значения
data = data.drop(columns="Unnamed: 0") # Удалим лишний столбец
data["timestamp"] = pd.to_datetime(data["timestamp"]) # Переделаем время в тип datetime
```

```python
data
```

```python
gender = {"М": True, "Ж": False} 
data["gender"] = data["gender"].map(gender) # Заменим строки пола на bool значения (занимают меньше памяти)
```

```python
data
```

```python
data.describe()
```

```python
data_confirmation = data[data["action"] == "confirmation"]
data_confirmation
```

Подсчёт общего объёма продажи

```python
income_per_year = data_confirmation.drop(columns=["userid", "sessionid", "action", "category", "age", "gender", "city"])
income_per_year.set_index("timestamp", inplace=True)
income_per_year = income_per_year.resample("YE").sum()
```

```python
income_per_year["value"].apply(lambda x: f'{x / 1e6:.2f}M')
```
